```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util

# 1. Load dataset
df = pd.read_excel('clean_QA.xlsx')

# 2. Load embedding model (no training needed)
model = SentenceTransformer('all-MiniLM-L6-v2')

# 3. Compute embeddings for all existing questions
question_embeddings = model.encode(df['question'].tolist(), convert_to_tensor=True)

def chatbot(question):
    # Encode user question
    query_emb = model.encode(question, convert_to_tensor=True)

    # Compute similarity
    scores = util.cos_sim(query_emb, question_embeddings)[0]

    # Find highest score
    best_idx = scores.argmax().item()
    best_score = scores[best_idx].item()

    # If similarity too low → say "I don't know"
    if best_score < 0.45:
        return "Sorry, I don't have information about that."

    return df.iloc[best_idx]["answer"]

# Test
while True:
    q = input("Ask me a health question: ")
    print("Answer:", chatbot(q))
    print()
```
